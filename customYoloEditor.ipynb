```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import os
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/best.pt', force_reload=True)
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture('twc.mp4')
# Define the output video writer
output_file = 'output11.mp4'
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'MP4V') 
output_video = cv2.VideoWriter(output_file, fourcc, fps, (int(cap.get(3)), int(cap.get(4))))

timings = []  # List to store the timings of frames with a face

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    results = model(frame)
    #print(results.pred)
    if len(results.pred[0]) > 0:
        # Face is detected
        frame_number = cap.get(cv2.CAP_PROP_POS_FRAMES)
        frame_timing_sec = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

        timings.append((frame_number, frame_timing_sec))

        # Write the frame with object detection to the output video
    rendered_frame = np.squeeze(results.render())
    output_video.write(rendered_frame)

    #cv2.imshow(rendered_frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
output_video.release()
cv2.destroyAllWindows()

# Print the frame number and timing in seconds of frames with a face
# for frame_number, frame_timing_sec in timings:
#     print(f"Face detected at frame {frame_number}, timing {frame_timing_sec} seconds")

```
